```python
import datetime, requests, yfinance
from alpaca_trade_api.rest import REST

BASE_URL = 'https://paper-api.alpaca.markets'
API_KEY = input("Enter your API Key: ")
SECRET_KEY = input("Enter your Secret Key: ")

api = REST(key_id=API_KEY, secret_key=SECRET_KEY, base_url=BASE_URL)
```

```python
#Get news for symbol
news = api.get_news('TWTR')
news
```

```python
#specify timeframe for news
news = api.get_news('TWTR', start='2022-06-01', end='2022-10-31', include_content = True) #include_content outputs the body of the article right to the user
news
```

```python
#summarize the information from Benzinga API to more readable format by iterating through
for story in news:
  print(story.headline)
  print(story.summary)
  for image in story.images:
    print(image['url'])
```

```python
#complete sentiment analysis on returned news articles associated with a symbol with transformers machine learning
#https://huggingface.co/docs/transformers/main_classes/pipelines
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
```

```python
#example sentiment analysis
classifier('happy')
```

```python
classifier('murder')
```

```python
#apply the sentiment analysis to the news headlines related to the ticker
for story in news:
  print(story.headline)
  print(classifier(story.headline))
```
